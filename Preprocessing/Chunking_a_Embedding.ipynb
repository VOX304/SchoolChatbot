```python
from sentence_transformers import SentenceTransformer

class CustomEmbedding:
    def __init__(self, model_name):
        self.model = SentenceTransformer(model_name, trust_remote_code=True)  # ✅ Add trust_remote_code=True

    def embed_documents(self, texts):
        return self.model.encode(texts, convert_to_numpy=True)

    def embed_query(self, text):
        return self.model.encode([text], convert_to_numpy=True)[0]
    def __call__(self, text):
        """Make the class callable, so FAISS can use it."""
        return self.embed_query(text)

# Initialize the embedding model
Viet_embedding_model = CustomEmbedding("dangvantuan/vietnamese-document-embedding")

# Generate embeddings

```

# Load and Organize

```python
import os
import json

def load_json_from_folder(folder_path):
    """Reads all .json files in a folder and returns a list of dictionaries."""
    json_data = []

    for filename in os.listdir(folder_path):
        if filename.endswith(".json"):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, "r", encoding="utf-8") as file:
                try:
                    data = json.load(file)
                    if isinstance(data, list):
                        json_data.extend(data)  # Ensure proper structure (list of dicts)
                    else:
                        json_data.append(data)
                except json.JSONDecodeError:
                    print(f"Warning: Failed to parse {filename}, skipping.")

    return json_data


def organize_metadata_by_document(data):
    """Step 1: Organize metadata by document_name."""
    document_chunks = {}

    for item in data:
        doc_name = item.get('document_name')
        if not doc_name:
            continue  # Skip items without a document name

        if doc_name not in document_chunks:
            document_chunks[doc_name] = {'text': [], 'table': [], 'image': []}

        datatype = item.get('datatype')
        extracted_content = item.get('extracted_content')

        if datatype == 'txt' and extracted_content:
            document_chunks[doc_name]['text'].append(extracted_content)

        elif datatype == 'table' and extracted_content:
            document_chunks[doc_name]['table'].append(extracted_content)

        elif datatype == 'img':
            img_data = {
                'img_path': item.get('img_path'),
                'extracted_content': extracted_content
            }
            document_chunks[doc_name]['image'].append(img_data)

    return document_chunks


# Example usage:
```

# Chunking

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document

def process_metadata_with_chunking(organized_metadata, recursive_chunk_size=512, recursive_chunk_overlap=50):
    """
    Process metadata by chunking text data using recursive chunking while preserving tables and images.

    Args:
    - organized_metadata: Dictionary containing document metadata.
    - recursive_chunk_size: The maximum size of each recursive chunk.
    - recursive_chunk_overlap: The overlap between consecutive recursive chunks.

    Returns:
    - updated_metadata: Dictionary with chunked text data and preserved table and image data.
    """
    updated_metadata = {}

    # Initialize the RecursiveCharacterTextSplitter
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=recursive_chunk_size, chunk_overlap=recursive_chunk_overlap)

    # Iterate through each document in the organized metadata
    for doc_name, content in organized_metadata.items():
        doc_data = {'text': [], 'table': [], 'image': []}

        # 1. Apply recursive chunking to the document's text
        if content['text']:
            # Prepare the text data in the correct format for the text splitter (as Document instances)
            documents = [Document(page_content=doc_text) for doc_text in content['text']]

            # Split the text into recursive chunks
            recursive_chunks = text_splitter.split_documents(documents)  # Now the format is correct

            doc_data['text'] = [chunk.page_content for chunk in recursive_chunks]  # Extract 'page_content'
            print(f"Recursive chunking completed for {doc_name} with {len(doc_data['text'])} chunks.")
        else:
            print(f"Skipping chunking for {doc_name} because no text data found.")

        # 2. Add tables without modification
        if content['table']:
            doc_data['table'] = content['table']

        # 3. Add images without modification
        if content['image']:
            doc_data['image'] = content['image']

        # Store the processed content under the respective document name
        updated_metadata[doc_name] = doc_data

    return updated_metadata
```

```python
folder_path = "/content/sample_data/TCU"  # Set your actual folder path

# Step 1: Load all JSON files
json_files_data = load_json_from_folder(folder_path)

# Step 2: Organize metadata by document name
organized_metadata = organize_metadata_by_document(json_files_data)
```

```python
chunking = process_metadata_with_chunking(organized_metadata)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
from langchain.vectorstores import FAISS

# Custom embedding class for text
# Perform the chunking

  # Prepare data for embedding
chunks = []
image_embeddings = []
table_embeddings = []


# Extracting and embedding text content from images, tables, and chunks
for doc_name, doc_data in chunking.items():
    # Embed text chunks
    if doc_data['text']:
        text_chunks = doc_data['text']
        text_embeddings = Viet_embedding_model.embed_documents(text_chunks)
        chunks.extend(text_chunks)  # Text chunks for FAISS

    # Embed extracted content from images (which is already text)
    if doc_data['image']:
        for image_data in doc_data['image']:
            extracted_content = image_data.get('extracted_content', '')  # Extracted text from the image
            if extracted_content:
                image_embedding = Viet_embedding_model.embed_query(extracted_content)
                image_embeddings.append(image_embedding)  # Store the embedding for extracted content

    # Embed tables (assuming table data is in text format)
    if doc_data['table']:
        for table in doc_data['table']:
                table_embeddings.append(Viet_embedding_model.embed_documents([table])[0])  # Embedding for tables


# Combine all the embeddings (text, image, and table)
all_embeddings = np.concatenate([text_embeddings, image_embeddings, table_embeddings], axis=0)

# Create a FAISS vector store for the embeddings
vector_db = FAISS.from_documents(chunks + [str(i) for i in range(len(image_embeddings))] + [str(i) for i in range(len(table_embeddings))], embedding_model=Viet_embedding_model)

# Save the FAISS vector store locally
vector_db.save_local("FaissVecDB")

print("FAISS vector store has been saved locally.")
```
