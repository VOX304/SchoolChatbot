```python
import fitz  # PyMuPDF

def extract_text_pymupdf(pdf_files):
    for pdf in pdf_files:
        print("Extracting with PyMuPDF: \n")
        doc = fitz.open(pdf)
        text = ""
        for page in doc:
            text += page.get_text("text")  # Extract as raw text
        print(text[:1000])  # Print first 1000 characters
        print("\n" + "="*80 + "\n")

extract_text_pymupdf(pdf_test_file)
```

```python
import pdfplumber

def extract_text_pdfplumber(pdf_files):
    for pdf in pdf_files:
        print("Extracting with pdfplumber: \n")
        text = ""
        with pdfplumber.open(pdf) as pdf_reader:
            for page in pdf_reader.pages:
                text += page.extract_text() or ""  # Avoid NoneType errors
        print(text[:1000])  # Print first 1000 characters
        print("\n" + "="*80 + "\n")

extract_text_pdfplumber(pdf_test_file)
```

```python
from markitdown import MarkItDown

def extract_text_markitdown(pdf_files):
    for pdf in pdf_files:
        print(f"Extracting with MarkItDown: {pdf}\n")

        md = MarkItDown(enable_plugins=False)  # Set to True if you want extra parsing
        result = md.convert(pdf)  # Returns a DocumentConverterResult object


        print(result)  # Print first 1000 characters
        print("\n" + "="*80 + "\n")

extract_text_markitdown(pdf_test_file)
```

```python
import fitz  # PyMuPDF
import cv2
import os
import numpy as np
import torch
from PIL import Image
import matplotlib.pyplot as plt
from paddleocr import PaddleOCR, draw_ocr
from vietocr.tool.predictor import Predictor
from vietocr.tool.config import Cfg

# Load PaddleOCR for Vietnamese text detection
paddle_ocr = PaddleOCR(lang="vi", use_angle_cls=True)

# Load VietOCR for better text recognition
config = Cfg.load_config_from_name('vgg_transformer')
config['device'] = 'cuda' if torch.cuda.is_available() else 'cpu'  # Use GPU if available
vietocr_detector = Predictor(config)
```

```python

# Open the PDF document
doc = fitz.open(pdf_test_file[1])
output_text = ""

# Create an output directory for images
os.makedirs("pdf_pages", exist_ok=True)
```

```python


# Iterate over all pages
for page_num in range(doc.page_count):
    # Load the page
    page = doc.load_page(page_num)

    # Render page to an image
    pix = page.get_pixmap()
    image_path = f"pdf_pages/page-{page_num + 1}.png"
    pix.save(image_path)  # Save page as an image

    # Load image for OCR
    image = cv2.imread(image_path)

    # Perform OCR with PaddleOCR
    results = paddle_ocr.ocr(image_path, cls=True)

    cropped_images = []  # Store cropped text regions

    for line in results[0]:  # Process detected text lines
        if isinstance(line, list) and len(line) == 2:
            box, text_info = line
            if isinstance(box, list) and len(box) == 4 and isinstance(text_info, tuple) and len(text_info) == 2:
                text, confidence = text_info

                # Get bounding box coordinates
                x_min = min(point[0] for point in box)
                y_min = min(point[1] for point in box)
                x_max = max(point[0] for point in box)
                y_max = max(point[1] for point in box)

                # Validate and crop the image
                if 0 <= x_min < x_max < image.shape[1] and 0 <= y_min < y_max < image.shape[0]:
                    cropped = image[int(y_min):int(y_max), int(x_min):int(x_max)]
                    cropped_images.append(Image.fromarray(cropped))

    # Process cropped images with VietOCR
    for img in cropped_images:
        recognized_text = vietocr_detector.predict(img)
        output_text += recognized_text + "\n"

# Save final extracted text
with open("final_extracted_text.txt", "w", encoding="utf-8") as file:
    file.write(output_text)

print("OCR process completed! Extracted text saved to 'final_extracted_text.txt'")
```
